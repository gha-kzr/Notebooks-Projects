```python
import pandas as pd
```

```python
# Fonctions utiles pour la suite du notebook

def lower_and_split_words(x):
    return str(x).lower().split(" ")

def check_skills(sr, skill):
  pos = sr[sr == skill]
  return pos
```

```python
# Thanks to ask9 - arbazkhan971 for the dataset provided on kaggle
# https://www.kaggle.com/datasets/arbazkhan971/allskillandnonskill

# On ouvre le fichier et on supprime les lignes incomplètes
filepath = "/content/drive/MyDrive/Colab Notebooks/Projects/user-languages.csv"
dataset = pd.read_csv(filepath, sep =';', header=None)
skills_text = dataset.iloc[0, 0].split(',')

skills = pd.Series(skills_text, name='skills')
skills = skills.str.lower()

skills = skills.drop_duplicates()
print(skills.describe())
```

```python
# On sauvegarde notre dataset dans un nouveau fichier pour l'utiliser dans l'extraction des compétences dans les descriptions des offres d'emploi
savepath = "/content/drive/MyDrive/Colab Notebooks/Projects/cleaned_skills_small.csv"
skills.to_csv(savepath, sep =',', index=False)
```
